### 데이터 (Data)
- 정보를 나타내는 숫자, 문자, 기호의 집합
- 가공되지 않은 사실이나 값의 집합

### 데이터베이스 (Database)
- 체계적으로 구성된 데이터의 집합
- 특정 목적을 위해 조직화되고 관리
- 정보를 효율적으로 저장, 검색, 업데이트, 관리할 수 있는 기술적인 도구와 구조를 제공

### 데이터베이스 관리 시스템 (DBMS; Database Management System)
- 데이터베이스를 관리할 수 있는 기술적인 소프트웨어

### 관계형 데이터베이스 시스템 (Relational DBMS)
- 데이터를 테이블(표) 형식으로 관리하고, 각 테이블 간의 관계를 설정해서 데이터를 효율적으로 저장하고 검색하는 데이터베이스 관리 시스템의 한 종류
- SQL (Structed Query Language)를 사용 : 관계형 데이터베이스에서 데이터를 정의하고 조작하기 위한 언어
- 단점 : 테이블끼리 관계성을 맺고 무결성을 체크해야하므로 속도가 느림

### MySQL
- 설치 링크 : https://dev.mysql.com/downloads/installer/
- 원래 서버 컴퓨터에 설치하는 것이 맞음
- 웹 기준 설명
    1. www.naver.com(도메인)을 입력하게 되면 도메인을 관리하는 서버로 가서
    2. 네이버와 매핑되어있는 ip주소를 알려줌
    3. ip주소를 받아서 특정 컴퓨터를 찾아가서 웹 서버에 페이지 요청
    4. 웹 서버 안의 초기 화면을 브라우저로 응답
    5. 웹 서버 : 화면을 보내주는 역할.
    6. 정보를 저장하거나 데이터를 저장하는 것은 웹 서버가 데이터베이스와 통신하여 주고 받음
    7. DB와 웹 서버가 같이 있을 수도, 따로 있을 수도 있음
    8. MySQL은 웹 서버 컴퓨터에 깔아서 사용함
