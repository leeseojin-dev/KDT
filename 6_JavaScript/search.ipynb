```python
# 선형 탐색
def linear_search(arr, want):
    for x in range(len(arr)):
        if want == arr[x]:
            return x
    return -1
        
arr = [4, 5, 2, 14, 86, 34]
print(linear_search(arr, 34))
```

```python
# 선형 탐색
# enumerate()를 사용하여 더 파이썬스럽게
def linear_search(arr, want):
    for i, val in enumerate(arr):
        if val == want:
            return i
    return -1
        
arr = [4, 5, 2, 14, 86, 34]
print(linear_search(arr, 34))
```

```python
#이진 탐색
def binary_search(arr, want):
    left = 0
    right = len(arr) - 1

    while left <= right: 
        mid = (left + right) // 2
        if arr[mid] == want:
            return mid
        elif arr[mid] < want:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [2, 4, 5, 14, 34, 86]
print(binary_search(arr, 34))

```
