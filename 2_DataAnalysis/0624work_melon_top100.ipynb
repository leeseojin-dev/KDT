```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


# 서버 주소
url = "https://www.melon.com/chart/index.htm"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# 서버 요청
response = requests.get(url, headers=headers)

# HTML 가져오기
html = response.text

# HTML 분석
soup = BeautifulSoup(html, "html.parser")

rows = soup.select("tbody tr")

data = []


for row in rows:
    rank_tag = row.select_one("span.rank")  # 순위 가져오기
    title_tag = row.select_one("div.ellipsis.rank01")   # 곡 제목 가져오기
    singer_tag = row.select_one("div.ellipsis.rank02")  # 가수명 가져오기
    album_tag = row.select_one("div.ellipsis.rank03")  # 앨범명 가져오기
    likecnt_tag = row.select_one("span.cnt")  # 좋아요 수 가져오기

    if not title_tag:
        continue

    rank = rank_tag.get_text(strip=True)
    title = title_tag.get_text(strip=True)
    singer = singer_tag.get_text(strip=True)
    album = album_tag.get_text(strip=True)
    likecnt = likecnt_tag.get_text(strip=True)

    data.append({
            "순위": rank,
            "곡 제목": title,
            "가수명": singer,
            "앨범명": album,
            "좋아요 수": likecnt
        })


df = pd.DataFrame(data)
df.index = df.index + 1

file_name = f"0624work_melon_top100.csv"
df.to_csv(file_name, encoding="utf-8-sig")

df
```
