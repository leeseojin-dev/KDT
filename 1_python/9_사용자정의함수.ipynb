# **1. 사용자 정의 함수**

사용자 정의 함수란 사용자가 특정 작업을 수행하기 위해 직접 작성한 함수를 의미합니다. 파이썬에는 많은 내장 함수들이 있지만, 때로는 우리의 요구사항에 맞게 동작하는 함수를 직접 만들어야 할 때가 있습니다. 이때 사용자 정의 함수를 작성하게 됩니다.

### 1. 함수의 동작

```
def 함수명(매개변수1, 매개변수2, ...):
    # 함수 내용
    return 결과값
```

이름의 함수를 만드는 것이 아니라, 정확히는 다음처럼 동작합니다.
1. 함수 객체를 메모리에 생성
2. 그 함수 객체에 함수명 이름을 붙임

````python
def hello():
    print("안녕하세요")

a = hello
a()
hello()

# 👉 a = hello는 함수를 복사한 것이 아니라, 같은 함수 객체를 참조하는 이름을 하나 더 만든 것입니다.
````

### 2. 함수 객체

```
function object
├── 함수 이름
├── 매개변수 정보
├── 실행할 코드 정보

├── 전역 네임스페이스 정보
└── 클로저 정보 : 함수 안에 함수가 있는 경우
```

### 3. 함수 호출 시 메모리에서 일어나는 일⭐

2. 함수 실행을 위한 '새 공간' 생성
3. 매개변수 a, b에 값 연결
4. 함수 본문 실행
5. return 값 반환
6. 함수 실행 공간 제거

> 👉 이때 만들어지는 '새 공간'을 스택 프레임(stack frame) 또는 함수 호출 프레임이라고 합니다.

````python
print(a) # 함수 객체: 함수 자체를 나타내는 객체
print(a.__code__) # 코드 객체: 함수 안에서 실행할 실제 바이트코드 정보를 담고 있는 객체
````

### 함수 호출 프레임
함수 호출 프레임(call frame)은 파이썬에서 함수가 호출될 때마다 생성되는 독립적인 실행 공간으로, 해당 함수의 매개변수, 지역 변수, 반환 주소, 그리고 실행 상태(어디까지 실행했는지 등)를 저장하는 구조입니다. 이 프레임은 **스택(stack)** 형태로 관리되어 함수가 호출될 때 쌓이고, 실행이 끝나면 제거되며, 이를 통해 함수 간 변수 충돌 없이 안전하게 실행됩니다. 즉, 같은 함수를 여러 번 호출하더라도 매번 새로운 프레임이 만들어지기 때문에 각 호출은 서로 영향을 주지 않고 독립적으로 동작합니다.

### 2. 매개변수가 있고 반환 값이 없는 함수

````python
def func2(num):
    print(f'입력받은 숫자: {num}')
    
func2(10)
func2(4)
````

````python
def func3(start, end):
    sum = 0
    for i in range(start, end+1):
        sum += i
    print(f'{start}부터 {end}까지의 합: {sum}')

func3(1, 10)
func3(1, 100)
````

# **3. 변수의 범위**

변수의 범위(scope)는 해당 변수가 프로그램 내에서 참조되고 변경될 수 있는 영역을 의미합니다.

````python
def local_example():
    local_var = "로컬 변수"
    print(local_var)

local_example()

# local_var는 stack에 만들어지기 때문에 밖에서 사용을 할 수 없음
````

> 👉 지역변수: 가장 내부의 범위로, 함수 내에서 정의된 변수들이 해당됩니다. 함수 내에서만 접근할 수 있습니다.

````python
global_var = "I'm a global variable"

def test_global_scope():
    print(global_var)

test_global_scope()
````

> 👉  전역변수: 스크립트 전체에서 사용되는 범위로, 함수 외부에 정의된 변수가 해당됩니다. 모든 함수에서 접근할 수 있지만, 함수 내부에서 수정하려면 global 키워드가 필요합니다.

````python
global_var = 10

def modify_global():
    global_var = 20     # stack영역의 local변수
    print(global_var)       # 20

modify_global()
print(global_var)  # 10
````

````python
global_var = 10

def modify_global():
    global global_var
    global_var = 20

modify_global()
print(global_var)  # 20
````

````python
# 중첩 함수
def outer_function():
    enclosing_var = "둘러싼 범위 변수"
    
    def inner_function():
        print(enclosing_var)
    
    return inner_function       # 함수 자체를 return 
````

````python
f = outer_function()
f()
````

> 👉 클로저 변수: 내부 함수가 외부 함수의 지역 변수를 사용하면, 파이썬은 그 값을 바로 삭제하지 않고 **클로저 공간**에 보관합니다.

### 클로저 공간

클로저 공간은 내부 함수가 외부 함수의 변수를 계속 사용할 수 있도록, 그 값을 따로 저장해두는 메모리 영역(정확히는 객체 구조)입니다. 일반적으로 함수가 끝나면 그 안의 지역 변수는 사라지지만, 내부 함수가 그 변수를 참조하고 있으면 파이썬은 해당 값을 지우지 않고 함수 객체 안에 __closure__라는 형태로 보관합니다. 이때 저장되는 값은 보통 cell 객체로 감싸져 유지되며, 내부 함수는 실행될 때 이 클로저 공간을 통해 외부 변수에 접근합니다.

> heap영역에 클로저 객체를 저장하는 것

```
len = 5  # 내장 함수 'len'을 덮어씀
print(len([1, 2, 3]))

# TypeError: 'int' object is not callable
# 오류 발생! len이 함수가 아니라 변수로 인식됨 
```

> 👉 빌트인 함수: 파이썬에서 별도의 import 없이 바로 사용할 수 있도록 기본적으로 제공되는 함수들을 의미합니다. 예를 들어 print(), len(), type(), range() 같은 함수들이 이에 해당하며, 파이썬이 실행될 때 자동으로 Built-in 네임스페이스에 등록되어 있기 때문에 어디서든 이름만으로 호출할 수 있습니다. 변수 이름을 찾을 때 적용되는 LEGB 규칙(Local → Enclosing → Global → Built-in)에서 마지막 단계에 해당하며, 만약 같은 이름의 변수를 직접 정의하면 Built-in 함수는 가려질 수 있습니다.

# **5. 콜백함수**⭐

콜백 함수(callback function)는 다른 함수에 인자로 전달되어, 특정 시점이나 조건이 되었을 때 나중에 호출되는 함수를 의미합니다. 파이썬에서는 함수도 객체이기 때문에 변수처럼 전달할 수 있으며, 이를 활용해 어떤 작업이 끝난 뒤 실행할 동작을 미리 정의해 둘 수 있습니다. 즉, 콜백 함수는 “지금 바로 실행하는 함수”가 아니라, 다른 함수에게 맡겨두었다가 필요할 때 호출되도록 하는 함수입니다.
> 개발자가 호출한것이 아닌 함수가 다른 함수를 호출하는 것

````python
def say_hello(name):
    print(f"{name}님 안녕하세요!")

def execute(callback):      # 여기 callback은 변수명 (a등 다른 이름 사용해도 됨)
    print("작업을 시작합니다.")
    callback("김사과")      # 여기 callback()은 함수명. "김사과"는 매개변수. 
    print("작업을 종료합니다.")

execute(say_hello)      # 👉여기서 say_hello가 콜백 함수입니다.

# say_hello() : 지금 당장 함수를 실행해라
# say_hello : 함수의 주소
# -> execute에 say_hello의 함수 주소를 넘기는 것
````

***

````python
def print_score(score):
    print(f"점수: {score}")

def print_pass_fail(score):
    if score >= 80:
        print(f"{score}점: 합격")
    else:
        print(f"{score}점: 불합격")

def print_bonus_score(score):
    print(f"보너스 적용 점수: {score + 5}")

def process_scores(scores, callback):
    for score in scores:
        callback(score)
````

````python

scores = [80, 95, 70, 100, 60]

print("원점수 출력")
process_scores(scores, print_score)

print("합격 여부 출력")
process_scores(scores, print_pass_fail)

print("보너스 점수 출력")
process_scores(scores, print_bonus_score)
````

> 👉 콜백 함수는 공통 작업은 그대로 두고, 세부 동작만 바꾸고 싶을 때 유용합니다.

***

````python
students = [
    {"name": "김사과", "score": 90},
    {"name": "반하나", "score": 75},
    {"name": "오렌지", "score": 100}
]

def get_score(student):
    # print(student["score"], end = " ") # 90 75 100
    return student["score"]
    
````

````python
result = sorted(students, key=get_score)
print(result)

# sorted() : 컬렉션 정렬해서 result에 저장 (원본은 유지)
# key=get_score :  key 뒤에 있는 것(get_score)으로 sorted할거야
# get_score = 콜백함수

# 후에 굉장히 많은 데이터 처리를 위해 콜백함수 많이 사용
````

> 👉 sorted()는 학생 하나하나를 비교할 때마다 get_score()를 호출해서 정렬 기준값을 얻습니다.

# **6. 람다 함수**

람다 함수(lambda function)는 파이썬에서 이름 없이 간단한 함수를 한 줄로 정의할 수 있는 익명 함수로, 주로 짧고 일회성으로 사용되는 연산을 표현할 때 활용됩니다. 일반 함수(def)에 비해 구조는 제한적이지만 코드가 간결해지고, 특히 sorted, map, filter 같은 함수에서 콜백 함수로 자주 사용됩니다.
람다는 무조건 return형. 이름이 없는 함수이기 때문에 일반적인 함수 사용 방법(함수명으로 호출)으로는 사용이 불가능. 

```
lambda 매개변수: 반환할_표현식
```

````python
# 일반 함수
def square(x):
    return x * x
print(square(5))

# 람다 함수
square = lambda x: x * x
print(square(5))

# 람다 함수 여러 개의 매개 변수
multiply = lambda a, b: a * b
print(multiply(3, 4))
````

````python
pass_fail = lambda score: "합격" if score >= 60 else "불합격"

print(pass_fail(80))
print(pass_fail(45))
````

````python
students = [
    {"name": "김사과", "score": 90},
    {"name": "반하나", "score": 75},
    {"name": "오렌지", "score": 100}
]

# def get_score(student):
#     return student["score"]

result = sorted(students, key=lambda student: student["score"])
print(result)

# 위의 예제를 lambda를 사용하면 함수를 메모리에 올리지 않고 사용이 가능하다
````

````python
# 👉 map()은 리스트의 각 요소에 같은 작업을 적용할 때 사용합니다

scores = [70, 80, 90]

result = list(map(lambda score: score + 10, scores))    
print(result)

# list함수로 감싸야 map함수 안의 데이터 출력 가능. (map객체, filter객체를 반환하기 때문에)
````

````python
# 👉 filter()는 조건을 만족하는 값만 골라낼 때 사용합니다.

scores = [45, 70, 88, 52, 100]

passed = list(filter(lambda score: score >= 60, scores))
print(passed)
````

***
- 함수를 사용하는 이유는 반복된 과정을 재사용하기 위함

- 변수 이름을 찾을 때 적용되는 LEGB규칙 (Local -> Enclosing -> Global -> Built-in) 때문에 변수 범위 수업 시 지역 변수가 우선시라는 것을 여기서도 알 수 있다.
- 함수 정의 시 함수 객체가 heap에 올라가고 호출(call)되면 stack에 복사돼서 사용
- 콜백 함수 : 본인이 아닌 다른 함수에 의해 호출되는 함수
    - 잘 사용하면 굉장히 확장성있게 사용 가능
- 간단한 콜백 함수는 람다 함수 사용하면 굳~
- 후에 굉장히 많은 데이터 처리를 위해 콜백함수 많이 사용하게 될 것임
- 람다 함수는 메모리에 올라가지 않는건가? -> False. 람다 함수는 실행 시 heap에 생성되고 참조가 사라지면 가비지 컬렉터(Garbage Collector)에 의해 즉시 메모리가 해제
