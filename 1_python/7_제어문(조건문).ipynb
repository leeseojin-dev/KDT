**파이썬의 조건문은 프로그램의 실행 흐름을 제어하기 위한 문법**으로, 기본적으로 if, elif, else를 사용하여 조건의 참(True)과 거짓(False)에 따라 다른 코드를 실행할 수 있습니다. 또한 간단한 조건에 대해서는 값1 if 조건 else 값2 형태의 조건부 표현식을 사용해 한 줄로 값을 선택할 수 있으며, Python 3.10부터는 match ~ case 문법을 활용한 구조적 패턴 매칭을 통해 값뿐만 아니라 리스트나 딕셔너리 같은 데이터의 구조를 기준으로 분기 처리가 가능합니다. 이처럼 파이썬은 상황에 따라 다양한 방식의 조건문을 제공하여 간단한 분기부터 복잡한 데이터 처리까지 유연하게 구현할 수 있습니다.

# **1. if, elif, else**

### 1. if문

가장 기본적인 조건문입니다. 특정 조건이 참(True)일 경우에만 해당 블록 내의 코드가 실행됩니다.

```
if 조건:
    실행할 코드
```
여기에서 조건은 'True 또는 False로 평가될 수 있는' 표현식이어야 합니다.<br>
조건이 True로 평가되면 if 문 바로 아래의 코드 블록이 실행됩니다. 그렇지 않으면 그 코드 블록은 건너뜁니다.

````python
if True:
    print('결과는 참입니다')
print('프로그램을 종료합니다')

if False:
    print('결과는 참입니다')    # 실행x
print('프로그램을 종료합니다')

if '안녕하세요':
    print('결과는 참입니다')
print('프로그램을 종료합니다')

if '':
    print('결과는 참입니다')    # 실행x
print('프로그램을 종료합니다')

if '안녕하세요':print('결과는 참입니다')
print('프로그램을 종료합니다')
````

````python
age = int(input('나이를 입력하세요: '))

if age > 19:
    print('성인입니다')

print('입력한 나이: ', age)
````

### 2. else문

else 문은 조건문에서 특정 조건이 충족되지 않을 때 실행할 코드 블록을 정의할 때 사용합니다.

```
if 조건:
    코드 블록1      # True
else:
    코드 블록2      # False
```

````python
age = int(input('나이를 입력하세요'))

if age > 19:
    print('성인입니다')
    print('넷플릭스 성인영화를 감상할 수 있어요')
else:
    print('미성년입니다')
    print('디즈니플러스를 보세요')
print('입력한 나이: ', age)
````

### 왈러스 연산자
:= 연산자는 파이썬에서 값을 할당하면서 동시에 그 값을 표현식 안에서 사용할 수 있게 해주는 할당 표현식(assignment expression)입니다. 기존에는 값을 변수에 저장한 뒤 다시 사용하는 두 단계 과정이 필요했지만, :=를 사용하면 조건문, 반복문, 컴프리헨션 등에서 값을 한 번 계산하면서 변수에 저장하고 바로 활용할 수 있어 코드가 더 간결해집니다.

### 3. elif 문

elif는 if-else 구조에서 추가적인 조건을 검사하기 위해 사용합니다. elif는 "else if"의 줄임말이며, 주어진 **여러 조건 중에서 하나만 참일 때** 실행되는 코드 블록을 정의합니다.

```
if 조건1:
    코드 블록1
elif 조건2:
    코드 블록2
elif 조건3:
    코드 블록3
else:
    코드 블록4
```

- elif는 if 없이 단독으로 사용될 수 없습니다.
- elif는 필요에 따라 여러 번 사용될 수 있습니다.
- else는 선택적입니다. 사용될 경우, 항상 if나 elif 뒤에 나타나야 합니다.

````python
age = int(input('나이를 입력하세요: '))

if age > 19:
    print('성인입니다')
elif age > 14:
    print('청소년입니다')
elif age > 6:
    print('어린이입니다')
elif age > 0:
    print('유아입니다')
# else:
#     print('입력을 확인하세요')

print('입력한 나이: ', age)
````

# **3. 조건부 표현식(Conditional Expression)**

조건부 표현식(Conditional Expression)은 조건의 참(True)과 거짓(False)에 따라 값을 선택하는 한 줄짜리 조건문으로, 간단한 분기 로직을 간결하게 표현할 때 유용합니다. 다만 조건이 복잡하거나 여러 줄의 처리가 필요한 경우에는 일반적인 if문을 사용하는 것이 가독성 측면에서 더 적절합니다.

```
조건이 참인 경우 값 if 조건 else 조건이 거짓인 경우 값
```

````python
# if~else로 작성
num = int(input('숫자를 입력하세요: '))
if num % 2 == 0:
    print('짝수')
else:
    print('홀수')
````

````python
# 조건부 표현식으로 작성
num = int(input('숫자를 입력하세요: '))
print('짝수') if num % 2 == 0 else print('홀수')
````

````python
# if~elif~else로 작성
age = int(input('나이를 입력하세요'))

if age > 19:
    print('성인입니다')
elif age > 14:
    print('청소년입니다')
elif age > 6:
    print('어린이입니다')
elif age > 0:
    print('유아입니다')
else:
    print('다시 입력해주세요')
print('입력한 나이: ', age)
````

# **4. 구조적 패턴 매칭**

파이썬에서 if-elif-else 구문을 좀 더 간결하고 읽기 쉽게 만든 새로운 분기문입니다. Java/C의 switch-case와 비슷하지만 훨씬 더 강력한 기능을 가집니다.

```
match 값:
    case 패턴1:
        실행할 코드1
    case 패턴2:
        실행할 코드2
    case _:
        기본 실행 코드 (default)
```

````python
month = int(input("월을 입력하세요 (1~12): "))

match month:
    case 1 | 3 | 5 | 7 | 8 | 10 | 12:
        print(f"{month}월은 31일까지 있습니다.")
    case 4 | 6 | 9 | 11:
        print(f"{month}월은 30일까지 있습니다.")
    case 2:
        print("2월은 28일 또는 윤년이면 29일까지 있습니다.")
    case _:
        print("잘못된 월입니다. 1~12 사이의 숫자를 입력해주세요.")
````

````python
user = ("김사과", 20) # tuple

match user:     # 기본형, 컬렉션형 모두 가능
    case (name, age) if age > 19:       # match 뒤의 형과 case 뒤의 형이 맞아야 됨
        print(f"{name}님은 성인입니다.")
    case (name, age) if age > 15:
        print(f"{name}님은 청소년입니다.")
    case (name, age) if age > 6:
        print(f"{name}님은 어린이입니다.")
    case _:
        print(f'{name}님은 유아입니다.')
````

````python
# scores = [95, 88, 76]     # 리스트
# scores = (95, 88, 76)       # 튜플 (아래 설명 때문에 첫번째 case문에 걸림. 만약 요소 개수가 다르면 건너 뜀)
scores = {"국어":95, "영어":88, "수학":76}        # 딕셔너리 (딕셔너리는 딕셔너리형으로)

match scores:
    case [korean, english, math]:       # 리스트
        print(f"1. 국어: {korean}, 영어: {english}, 수학: {math}")
    case (korean, _, math):     # 튜플
        print(f"2. 국어: {korean}, 수학: {math}")
    case {"국어": korean, "영어": english, "수학": math}:       # 딕셔너리
        print(f"3. 국어: {korean}, 영어: {english}, 수학: {math}")
````

> 👉 리스트와 튜플은 서로 다른 타입이지만, 구조적 패턴 매칭에서는 “시퀀스 패턴”으로 인해 비슷하게 매칭될 수 있습니다.<br>
> 👉 타입을 명확히 구분하려면 list(), tuple() 패턴을 사용해야 합니다.(리스트와 튜플)

````python
# scores = {"국어":95, "영어":88, "수학":76}
scores = [95, 88, 76]
# scores = (95, 88, 76)

match scores:
    # 타입 먼저 구분
    case list():
        match scores:
            case [korean, english, math]:
                print(f"리스트 → 국어: {korean}, 영어: {english}, 수학: {math}")

    case tuple():
        match scores:
            case (korean, _, math):
                print(f"튜플 → 국어: {korean}, 수학: {math}")

    case dict():
        match scores:
            case {"국어": korean, "영어": english, "수학": math}:
                print(f"딕셔너리 → 국어: {korean}, 영어: {english}, 수학: {math}")
````

***
- if-else문 : 조건의 범위가 있을 때 (ex. 성적 90~100점, 80~89점)
- match문 : 값이 딱 일치하는 것을 찾을 때 (ex. 1~12월)
